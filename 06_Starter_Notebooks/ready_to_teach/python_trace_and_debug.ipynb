## Guided repair
The function must ignore `None`, reject non-numeric values, and raise `ValueError` when no usable values remain.


```python
from numbers import Real

def safe_average(values):
    cleaned = []
    for value in values:
        if value is None:
            continue
        if not isinstance(value, Real):
            raise TypeError(f"non-numeric value: {value!r}")
        cleaned.append(float(value))
    if not cleaned:
        raise ValueError("no usable numeric values")
    return sum(cleaned) / len(cleaned)

assert safe_average([1, 2, None, 3]) == 2.0
try:
    safe_average([None])
except ValueError as exc:
    print("Expected:", exc)
else:
    raise AssertionError("ValueError was not raised")
```

```python
def linear_search(values, target):
    for index, value in enumerate(values):
        if value == target:
            return index
    return -1

assert linear_search([5, 2, 9], 9) == 2
assert linear_search([5, 2, 9], 7) == -1
print("search tests passed")
```
