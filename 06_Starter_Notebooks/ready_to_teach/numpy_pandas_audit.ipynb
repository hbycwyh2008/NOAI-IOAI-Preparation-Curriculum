```python
from __future__ import annotations
import hashlib, json, platform, random
from pathlib import Path
import numpy as np
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
print("Python:", platform.python_version())
print("Working directory:", Path.cwd())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

rng=np.random.default_rng(SEED)
df=pd.DataFrame({
 'record_id':[f'R{i:03d}' for i in range(120)],
 'group':rng.choice(list('ABC'),120),
 'age':rng.normal(16,1.5,120).round(1),
 'hours':rng.gamma(2,2,120).round(2),
 'device':rng.choice(['desktop','tablet','phone'],120),
 'target':rng.integers(0,2,120),
})
df.loc[rng.choice(df.index,8,replace=False),'hours']=np.nan
df=pd.concat([df,df.iloc[[3]]],ignore_index=True)
print(df.head())
```

```python
audit={
 'shape':df.shape,
 'duplicate_rows':int(df.duplicated().sum()),
 'duplicate_ids':int(df['record_id'].duplicated().sum()),
 'missing':df.isna().sum().to_dict(),
 'dtypes':df.dtypes.astype(str).to_dict(),
 'target_distribution':df['target'].value_counts(normalize=True).round(3).to_dict(),
}
print(json.dumps(audit,indent=2))
assert audit['duplicate_rows']==1
assert audit['missing']['hours']==8
```

```python
fig,axes=plt.subplots(1,2,figsize=(8,3))
df['hours'].plot(kind='hist',bins=15,ax=axes[0],title='Hours')
df.groupby('group')['target'].mean().plot(kind='bar',ax=axes[1],title='Target rate by group')
plt.tight_layout(); plt.show()
```
