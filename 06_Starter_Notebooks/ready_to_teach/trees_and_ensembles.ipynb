```python
from __future__ import annotations
import hashlib, json, platform, random
from pathlib import Path
import numpy as np
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
print("Python:", platform.python_version())
print("Working directory:", Path.cwd())
```

```python
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import f1_score

X,y=make_classification(n_samples=700,n_features=12,n_informative=6,n_redundant=2,weights=[.65,.35],flip_y=.05,random_state=SEED)
Xtr,Xva,ytr,yva=train_test_split(X,y,test_size=.3,stratify=y,random_state=SEED)
curves=[]
for depth in [1,2,3,5,8,None]:
    m=DecisionTreeClassifier(max_depth=depth,random_state=SEED).fit(Xtr,ytr)
    curves.append({"depth":str(depth),"train_f1":f1_score(ytr,m.predict(Xtr)),"val_f1":f1_score(yva,m.predict(Xva))})
print(pd.DataFrame(curves))
```

```python
models={
 "tree":DecisionTreeClassifier(max_depth=4,random_state=SEED),
 "forest":RandomForestClassifier(n_estimators=120,min_samples_leaf=3,random_state=SEED,n_jobs=1),
 "boosting":GradientBoostingClassifier(random_state=SEED),
}
comparison=[]
for name,m in models.items():
    m.fit(Xtr,ytr)
    comparison.append({"model":name,"train_f1":f1_score(ytr,m.predict(Xtr)),"val_f1":f1_score(yva,m.predict(Xva))})
comparison=pd.DataFrame(comparison).sort_values('val_f1',ascending=False)
print(comparison)
```
